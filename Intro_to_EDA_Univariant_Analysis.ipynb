# **Intro to EDA**
Exploratory Data Analysis (EDA) is the process of summarizing the main characteristics of a dataset, often using visual and quantitative methods. It helps identify patterns, detect outliers, and discover relationships between variables.

***Statistical measures*** are essential for understanding the distribution and characteristics of the data. The following examples showcase key statistical functions in Pandas.

```python
#importing packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Calculating the Mean

* The *.mean()* method calculates the average of the values in the series.
* This measure provides insight into the central tendency of the data.

```python
# Creating a pandas Series containing a list of random numbers.
random_num=pd.Series([23,25,25,26,23,10000000])
```

```python
#the mean is sensitive to outliers, such as 10000000 in this example.
random_num.mean()
```

Finding the Mode
* The .mode() method returns the value(s) that appear most frequently in the series.
* In this example, mode()[0] extracts the first mode if there are multiple modes.

```python
# Retrieving the most frequent value (mode) from the 'random_num' Series.
# In this case, it will return the first mode value if there are multiple modes.
random_num.mode()[0]
```

Calculating the Median
* The .median() method calculates the middle value when the data is sorted.
* Unlike the mean, the median is less affected by outliers and gives a better representation of the central value for skewed data.

```python
# Calculating the median value of the 'random_num' Series.
random_num.median()
```

Generating a Statistical Summary
* The .describe() method provides a summary of the data

```python
# This will provide the count, mean, standard deviation, minimum,
# 25th percentile (Q1), median (50th percentile or Q2), 75th percentile (Q3), and maximum values.
random_num.describe()
```

##**Univariante EDA**
In Univariate EDA, we focus on analyzing one variable at a time. Visualizations help to better understand distributions, central tendencies, and variability of numeric columns.

###**Creating a Histogram**
* A histogram visualizes the distribution of a numeric variable by dividing it into intervals (bins).
* Helps identify whether the data is skewed, uniform, or normally distributed.

Using matplotlib **.plot()** method

```python
penguins['body_mass_g'].plot(kind='hist')
```

Adding ***statistical lines (mean, median, mode)*** enhances the plot and understand how different measures of central tendency relate to the data's distribution.
Mean vs. Median:
* If the mean is greater than the median, the data is positively skewed.
* If the mean is less than the median, the data is negatively skewed.


```python
ax=penguins['body_mass_g'].plot(kind='hist')
ax.vlines(penguins['body_mass_g'].mean(),0,70,color='grey')
ax.vlines(penguins['body_mass_g'].median(),0,70,color='green')
ax.vlines(penguins['body_mass_g'].mode(),0,70,color='yellow')
```

###**Understanding Skewness**
Skewness is a measure of the asymmetry in the distribution of a dataset. It helps determine whether the data leans more to the left (negative skew) or to the right (positive skew), or if it's symmetrical (skewness ≈ 0).
* Positive Skew: The tail is longer on the right. Values are concentrated on the lower end.
* Negative Skew: The tail is longer on the left. Values are concentrated on the higher end.
* Symmetrical Data: Skewness is close to 0, indicating a normal-like distribution.

```python
penguins.columns
```

Creates a histogram to visualize the distribution of bill_depth_mm

```python
penguins['bill_depth_mm'].plot(kind='hist',bins=20,color='blue',alpha=0.5,edgecolor='red')
```

Calculating Skewness

```python
#A negative value indicates negative skewness
skewness=penguins['bill_depth_mm'].skew()
print(skewness)
```

```python
#A positive value indicates positive skewness
skewness=penguins['body_mass_g'].skew()
print(skewness)
```

###**Box Plot**
* helps to provide a summary of its distribution and identifying potential outliers.

* *Median Line*: The horizontal line inside the box represents the median (50th percentile).
* *Box*: The interquartile range (IQR), showing the middle 50% of the data (from the 25th percentile to the 75th percentile).
* *Whiskers*: Extend to the minimum and maximum values within 1.5 times the IQR from the box.
* *Outliers*: Individual points outside the whiskers, highlighting extreme values.




```python
penguins['bill_depth_mm'].plot(kind='box')
```

Grouped Box Plot

the below box plot illustrate the bill_depth_mm column, separated by each unique value in the species column.

```python
penguins.boxplot(column='bill_depth_mm',by='species')
```

### **Bar Plots**
* Best suited for showing counts or aggregated values for categories.

For the below EDA,the bar plot helps to

* Displays the frequency of observations for each island (e.g., Biscoe, Dream, and Torgersen in the Penguins dataset).
* Highlights which island has the most or least penguin data recorded.

```python
penguins.groupby('island')['island'].count()
```
