```python
# Load the diabetes dataset
diabetes_sklearn = load_diabetes()

# Convert the dataset to a DataFrame
diabetes_df = pd.DataFrame(data=diabetes_sklearn.data,
                           columns=diabetes_sklearn.feature_names)

# Add target variable to the DataFrame
diabetes_df['target'] = diabetes_sklearn.target

print(diabetes_df.head())

# Print the shape of the feature matrix and target vector
print("Shape of Sklearn Diabetes Data:", diabetes_df.shape)
```

```python
print(diabetes_df.describe())
```

```python
print(diabetes_df.info())
```

```python
print(diabetes_df.isnull().sum())
print(diabetes_df.duplicated().sum())
```

```python
sns.boxplot(data=diabetes_df, x='sex', y='age')
plt.title('Age Distribution by Sex')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()
```

```python
# Histogram of BMI
plt.hist(diabetes_df['bmi'], bins=20, color='skyblue', edgecolor='black')
plt.title('BMI Distribution')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot of BMI vs. Target
sns.scatterplot(data=diabetes_df, x='bmi', y='target', alpha=0.7)
plt.title('BMI vs. Disease Progression')
plt.xlabel('BMI')
plt.ylabel('Target (Disease Progression)')
plt.show()
```

```python
# Pairplot of selected features
sns.pairplot(diabetes_df[['age', 'bmi', 'bp', 'target']])
plt.suptitle('Pairplot of Age, BMI, BP, and Target', y=1.2)
plt.show()
```
